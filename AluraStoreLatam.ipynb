### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

def analisis_facturacion (df, columna):
    return df[columna].sum()

columna = 'Precio'

resultados = {
    'Tienda 1': analisis_facturacion(tienda, columna),
    'Tienda 2': analisis_facturacion(tienda2, columna),
    'Tienda 3': analisis_facturacion(tienda3, columna),
    'Tienda 4': analisis_facturacion(tienda4, columna)
}

df_resultados = pd.DataFrame.from_dict(resultados, orient='index', columns=['Total Facturación'])
print(df_resultados)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear el gráfico de barras
fig, ax = plt.subplots()
tiendas = list(resultados.keys())
sumas = list(resultados.values())

ax.bar(tiendas, sumas)
ax.set_ylabel(columna)
ax.set_title('Suma total de la columna "{}" por tienda'.format(columna))
plt.show()
```

En este caso la Tienda 1 tiene mayor ganancias y la tienda 4 menores ganancias

# 3. Calificación promedio de la tienda


```python
def promedio_calificacion(df, columna_calificacion):
    if columna_calificacion not in df.columns:
        print(f"❌ Verifica la columna: '{columna_calificacion}' no se encontró.")
        return None

    # Calculate the mean of the specified column
    return df[columna_calificacion].mean().round(2)

columna_calificacion = 'Calificación'
resultados = {
    'Tienda 1': promedio_calificacion(tienda, columna_calificacion),
    'Tienda 2': promedio_calificacion(tienda2, columna_calificacion),
    'Tienda 3': promedio_calificacion(tienda3, columna_calificacion),
    'Tienda 4': promedio_calificacion(tienda4, columna_calificacion)
}

df_resultados = pd.DataFrame.from_dict(resultados, orient='index', columns=['Promedio Calificación'])
print(df_resultados)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Contar la frecuencia de cada calificación en cada tienda
calificaciones = [1, 2, 3, 4, 5]  # Asegura el orden
conteo_t1 = tienda['Calificación'].value_counts().reindex(calificaciones, fill_value=0)
conteo_t2 = tienda2['Calificación'].value_counts().reindex(calificaciones, fill_value=0)
conteo_t3 = tienda3['Calificación'].value_counts().reindex(calificaciones, fill_value=0)
conteo_t4 = tienda4['Calificación'].value_counts().reindex(calificaciones, fill_value=0)

# Crear DataFrame combinando todo
df_conteo = pd.DataFrame({
    'Tienda 1': conteo_t1,
    'Tienda 2': conteo_t2,
    'Tienda 3': conteo_t3,
    'Tienda 4': conteo_t4
}, index=calificaciones) # Added the missing closing parenthesis here

# Crear gráfico de barras agrupadas
ax = df_conteo.plot(kind='bar', figsize=(10, 6), colormap='tab10', edgecolor='black')

# Títulos y etiquetas
plt.title('📊 Cantidad de Calificaciones por Tienda', fontsize=14)
plt.xlabel('Calificación (1 a 5)', fontsize=12)
plt.ylabel('Cantidad de veces', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Tiendas', fontsize=10, title_fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostrar los valores encima de cada barra
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=9)

plt.show()
```

La tienda 1 es la que tiene la mayor cantidad de mala calificación, seguida de la tienda 4 y la tienda 3 es la que tiene mayor cantidad de excelente calificación, seguida de la tienda 2

# 4. Productos más y menos vendidos

```python
def producto_mas_menos_comprado(df, nombre_tienda, columna_producto='Producto'):
    conteo = df[columna_producto].value_counts()

    # Obtener el más y menos comprado
    producto_mas = conteo.idxmax()
    veces_mas = conteo.max()

    producto_menos = conteo.idxmin()
    veces_menos = conteo.min()

    print(f"\n {nombre_tienda}")
    print(f" Producto más comprado: {producto_mas} ({veces_mas} veces)")
    print(f" Producto menos comprado: {producto_menos} ({veces_menos} vec{'es' if veces_menos > 1 else ''})")

    return conteo
conteo1 = producto_mas_menos_comprado(tienda, 'Tienda 1')
conteo2 = producto_mas_menos_comprado(tienda2, 'Tienda 2')
conteo3 = producto_mas_menos_comprado(tienda3, 'Tienda 3')
conteo4 = producto_mas_menos_comprado(tienda4, 'Tienda 4')

```

La tienda 2 tiene la mayor cantidad de veces compradas en un producto y la tienda 4 y la 1 tienen la menor cantidad de veces que un producto se compro, despues de la tienda 2.

# 5. Envío promedio por tienda

```python
def promedio_envio(df, nombre_tienda, columna_envio='Costo de envío'):
    if columna_envio not in df.columns:
        print(f" '{columna_envio}' no está en {nombre_tienda}")
        return None
    promedio = df[columna_envio].mean().round(2)
    print(f" {nombre_tienda} promedio S/ {promedio} en envíos.")
    return promedio

# Convertir en DataFrame para visualizar mejor
df_envios = pd.DataFrame.from_dict(promedios_envio, orient='index', columns=['Promedio Envío (S/)'])
print("\n Promedio de gasto en envíos por tienda:\n")
print(df_envios)
```

```python
import matplotlib.pyplot as plt

ax = df_envios.plot(kind='bar', legend=False, color='orange', edgecolor='black', figsize=(8,5))
plt.title('🚚 Promedio de Gasto en Envíos por Tienda', fontsize=14)
plt.ylabel('Soles (S/.)', fontsize=12)
plt.xlabel('Tienda', fontsize=12)
plt.xticks(rotation=0)

# Mostrar los valores encima de cada barra
for i, v in enumerate(df_envios['Promedio Envío (S/)']):
    plt.text(i, v + 0.1, f'S/ {v:.2f}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
